```python

class car:
    color = 'black'
    body = 'regular'
    gas_tank = 0 # litr
    max_speed = 200 # km/h
    position = 0
    state = 'stopped'
    speed = 0

    def start (self): # Функция start создает имитацию объекта 'self' и резервирует ему место внутри класса  
        print ('Wroom!')
        print ('inside class:', self)
        self.state = 'started' # это и есть как раз метод от атрибута state
    
        
    def start (self): # Cоздаем метод  'stopped'
        self.state = 'stopped' # это и есть как раз метод от атрибута state 
        
    def fill_gas_tank (self, gas): # Cоздаем метод  'заполненый топливный бак'
        self.gas_tank += gas #Создаем атрибут gas в методе топливный бак
        
    def accelerate (self, v): # Cоздаем метод  'заполненый топливный бак'
        if self.speed +v > self.max_speed:
            self.speed = self.max_speed
        else:
            self.speed += v #Создаем атрибут v в методе скорость
            
    def brak (self): 
        self.speed = 0 
    
    def go (self, h):
        self.position += self.speed * h 
        self.gas_tank -= 10*h
    
 
         
car_0 = car() # Теперь объект создан.
car_0.start()
car_1 = car()
car_2 = car()



print ('outside class:', car_0)
print ('outside class:', car_1)

print ('Цвет авомобиля_0:', car_0.color) #Смотрим атрибут автомобиля
print ('Цвет авомобиля_1 :', car_1.color) #Смотрим атрибут автомобиля

print ('Топливный бак авомобиля_0:', car_0.gas_tank) #Смотрим атрибут автомобиля
print ('Топливный бак авомобиля_1 :', car_1.gas_tank) #Смотрим атрибут автомобиля


car_1.color = 'green'
car_0.color = 'blue'
car_1.fill_gas_tank(30)
car_0.fill_gas_tank(40)


print ('Цвет авомобиля_0:', car_0.color) #Смотрим атрибут автомобиля
print ('Цвет авомобиля_1 :', car_1.color) #Смотрим атрибут автомобиля
print ('Бак авомобиля_0:', car_0.gas_tank) #Смотрим атрибут автомобиля
print ('Бак авомобиля_1 :', car_1.gas_tank) #Смотрим атрибут автомобиля

car_0.accelerate(70)
print (car_0.speed)

car_0.accelerate(90)
print (car_0.speed)

car_0.accelerate(110)
print (car_0.speed)

print ('Содержание словаря объекта car: ', car.__dict__)
car_0.start()
```
